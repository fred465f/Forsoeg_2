```python
# Functions
def find_max(x, y):
    # First max
    maks_y = max(y)
    maks_index = y.index(maks_y)
    maks_x = x[maks_index]
    return maks_x, maks_y, maks_index

def find_min(x, y):
    # First min
    min_y = min(y)
    min_index = y.index(min_y)
    min_x = x[min_index]
    return min_x, min_y, min_index

def find_ekstremums_punkter(x, y, period):
    min_x, min_y, min_index = find_min(x, y)
    maks_x, maks_y, maks_index = find_maks(x, y)
    
    # Num periods
    num_periods = int(x[-1] - x[0] // period)
    
    # Find mins
    mins = []
    if min_x < period:
        for i in range(1, num_periods):
            mins.append(min_x * num_periods * i)
    else:
        min_x_new = min_x
        count = 0
        while min_x_new > period and count <= 100:
            min_x_new -= period
            count += 1
        for i in range(1, num_periods):
            mins.append(min_x * num_periods * i)
        
```
